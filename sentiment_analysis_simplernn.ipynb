```python
docs = [
    'the weather is very cold today',
    'i am learning machine learning',
    'she loves reading books at night',
    'he went to the market to buy fruits',
    'the phone battery is almost empty',
    'we watched a movie yesterday',
    'coffee tastes better in the morning',
    'students are preparing for exams',
    'the train arrived late at the station',
    'she is cooking dinner for her family',
    'i enjoy playing football on weekends',
    'the internet is very slow today',
    'birds are flying in the sky',
    'he is writing a new blog post',
    'they are planning a trip to the mountains',
    'my laptop is working fine now',
    'the book on the table is mine',
    'we are waiting for the bus',
    'it is raining heavily outside',
    'she completed her homework on time'
]
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts(docs)
```

```python
len(tokenizer.word_index)
```

```python
sequences = tokenizer.texts_to_sequences(docs)
sequences
```

```python
from keras.utils import pad_sequences
sequences = pad_sequences(sequences,padding='post')
sequences
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding
model = Sequential()
model.add(Embedding(17,output_dim=2,input_length=5))

model.summary()
```

```python
model.compile('adam','accuracy')
```

```python
pred = model.predict(sequences)
print(pred)
```

```python
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN, Embedding, Flatten
```

```python
(X_train,y_train),(X_test,y_test) = imdb.load_data()
```

```python
X_train = pad_sequences(X_train,padding='post',maxlen=50)
X_test = pad_sequences(X_test,padding='post',maxlen=50)
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Embedding(input_dim=10000, output_dim=2, input_length=50))
model.add(SimpleRNN(32, return_sequences=False))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
history = model.fit(X_train, y_train,epochs=5,validation_data=(X_test,y_test))
```
